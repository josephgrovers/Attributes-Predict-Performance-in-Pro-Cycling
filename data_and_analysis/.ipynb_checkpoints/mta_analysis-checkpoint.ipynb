### CREATE DATAFRAMES OF FULL DATA FOR TARGET STATIONS

```python
# Lists below to use in functions below to pull data for
# January - June in the years of 2018, 2019, 2020
# Run cell to use lists later.

# MTA weeks to pull to create 2020 files
week_nums_2020 = [200104, 200111, 200118, 200125, 200201, 
                  200208, 200215, 200222, 200229, 200307, 
                  200314, 200321, 200328, 200404, 200411, 
                  200418, 200425, 200502, 200509, 200516, 
                  200523, 200530, 200606, 200613, 200620, 
                  200627]

# MTA weeks to pull to create 2019 files
week_nums_2019 = [190105, 190112, 190119, 190126, 190202, 
                  190209, 190216, 190223, 190302, 190309, 
                  190316, 190323, 190330, 190406, 190413, 
                  190420, 190427, 190504, 190511, 190518, 
                  190525, 190601, 190608, 190615, 190622, 
                  190629]

# MTA weeks to pull to create 2018 files
week_nums_2018 = [180106, 180113, 180120, 180127, 180203, 
                  180210, 180217, 180224, 180303, 180310, 
                  180317, 180324, 180331, 180407, 180414, 
                  180421, 180428, 180505, 180512, 180519, 
                  180526, 180602, 180609, 180616, 180623, 
                  180630]
```

```python
# create dataframes for stations Times Sq, Grand Central, Lexington,
# 34 St. Herald, St.Union, 59 St Columbus, and Penn Station.  
# Data will be for total Jan - June timeframe for each year.

stations_2020 = mta_data.get_yearly_data_station_only(week_nums_2020)
stations_2019 = mta_data.get_yearly_data_station_only(week_nums_2019)
stations_2018 = mta_data.get_yearly_data_station_only(week_nums_2018)
```

```python
# create dataframes for Fulton Station's specified lines  
# Data will be for total Jan - June timeframe for each year.

fulton_2020 = mta_data.get_yearly_data_fulton(week_nums_2020)
fulton_2019 = mta_data.get_yearly_data_fulton(week_nums_2019)
fulton_2018 = mta_data.get_yearly_data_fulton(week_nums_2018)
```

```python
# create dataframes for Chambers St Station's specified lines  
# Data will be for total late Dec - June timeframe for each year.

chambers_2020 = mta_data.get_yearly_data_chambers(week_nums_2020)
chambers_2019 = mta_data.get_yearly_data_chambers(week_nums_2019)
chambers_2018 = mta_data.get_yearly_data_chambers(week_nums_2018)
```

### COMBINE STATION DATAFRAMES FOR EACH YEAR

```python
# Combine dataframes from all stations by year; add year and month column

df_2020 = pd.concat([stations_2020,fulton_2020,chambers_2020])
df_2020['Month'] = df_2020['DATE'].dt.month
df_2020['Year'] = df_2020['DATE'].dt.year

df_2019 = pd.concat([stations_2019,fulton_2019,chambers_2019])
df_2019['Month'] = df_2019['DATE'].dt.month
df_2019['Year'] = df_2019['DATE'].dt.year

df_2018 = pd.concat([stations_2018,fulton_2018,chambers_2018])
df_2018['Month'] = df_2018['DATE'].dt.month
df_2018['Year'] = df_2018['DATE'].dt.year
```

### Identify & Clean Outliers for total traffic
 - Step 1: Plot to see where there are dramatic spikes
 - Step 2: Filter those that are well above average (<1_000_000)
 - Step 3: Determine way to replace number with strong estimate -- in this case that estimate is done by average of same day of week ('5')
 - Step 4: Replace outlier values
 - Step 5: Verify outliers gone


```python
# Step 1

plt.plot(df_2020.TotalTraffic);
```

```python
plt.plot(df_2019.TotalTraffic);
```

```python
plt.plot(df_2018.TotalTraffic);
```

```python
# Step 2
# Look at 2020 outliers and identify any commonalities
# Note index of all occurances are on weekday 5

df_2020[df_2020.TotalTraffic > 500_000]
```

```python
# Step 2
# Look at 2019 outliers and identify any commonalities
# Note index of occurances are on weekdays 5,4,1

df_2019[df_2019.TotalTraffic > 500_000]
```

```python
# Step 2
# Look at 2018 outliers and identify any commonalities
# Note index of all occurances are on weekday 5

df_2018[df_2018.TotalTraffic > 500_000]
```

```python
# Step 2
# Make a list of the index for the rows with the outliers

outlier_2020_index_wk5 = list(df_2020[
    ((df_2020.TotalTraffic) > 500_000) & 
    (df_2020.weekday == 5)].index)

outlier_2019_index_wk5 = list(df_2019[
    ((df_2019.TotalTraffic) > 500_000) & 
    (df_2019.weekday == 5)].index)

outlier_2019_index_wk4 = list(df_2019[
    ((df_2019.TotalTraffic) > 500_000) & 
    (df_2019.weekday == 4)].index)

outlier_2019_index_wk1 = list(df_2019[
    ((df_2019.TotalTraffic) > 500_000) & 
    (df_2019.weekday == 1)].index)

outlier_2018_index_wk5 = list(df_2018[
    ((df_2018.TotalTraffic) > 500_000) & 
    (df_2018.weekday == 5)].index)


print(outlier_2020_index_wk5)
print(outlier_2019_index_wk5)
print(outlier_2019_index_wk4)
print(outlier_2019_index_wk1)
print(outlier_2018_index_wk5)
```

```python
# Step 3

mean_2020_outlier_replace_wk5 = df_2020[
    ((df_2020.TotalTraffic) < 500_000) 
    & (df_2020.weekday == 5)].TotalTraffic.mean()

mean_2019_outlier_replace_wk5 = df_2019[
    ((df_2019.TotalTraffic) < 500_000) 
    & (df_2019.weekday == 5)].TotalTraffic.mean()

mean_2019_outlier_replace_wk4 = df_2019[
    ((df_2019.TotalTraffic) < 500_000) 
    & (df_2019.weekday == 4)].TotalTraffic.mean()

mean_2019_outlier_replace_wk1 = df_2019[
    ((df_2019.TotalTraffic) < 500_000) 
    & (df_2019.weekday == 1)].TotalTraffic.mean()

mean_2018_outlier_replace_wk5 = df_2020[
    ((df_2019.TotalTraffic) < 500_000) 
    & (df_2019.weekday == 5)].TotalTraffic.mean()
```

```python
# Step 4

for x in outlier_2020_index_wk5:
    df_2020.iloc[x,6] = mean_2020_outlier_replace_wk5
    
for x in outlier_2019_index_wk5:
    df_2019.iloc[x,6] = mean_2019_outlier_replace_wk5
    
for x in outlier_2019_index_wk4:
    df_2019.iloc[x,6] = mean_2019_outlier_replace_wk4

for x in outlier_2019_index_wk1:
    df_2019.iloc[x,6] = mean_2019_outlier_replace_wk1
    
for x in outlier_2018_index_wk5:
    df_2018.iloc[x,6] = mean_2018_outlier_replace_wk5
```

```python
# Verify outliers are gone

print(df_2020[df_2020.TotalTraffic > 500_000])
print(df_2020[df_2020.TotalTraffic > 500_000])
print(df_2020[df_2020.TotalTraffic > 500_000])
```

## ANALYSIS

```python
# Find total traffic for each year January - June

def yearly_traffic(year_data):
    return year_data.TotalTraffic.sum()

ttl_traffic_2020 = yearly_traffic(df_2020)
ttl_traffic_2019 = yearly_traffic(df_2019)
ttl_traffic_2018 = yearly_traffic(df_2018)
print(f'2020 Jan-June Traffic: {yearly_traffic(df_2020)}')
print(f'2019 Jan-June Traffic: {yearly_traffic(df_2019)}')
print(f'2018 Jan-June Traffic: {yearly_traffic(df_2018)}')
```

```python
# Find total traffic for each year April - June

Apr_June_2020 = month_2020[(month_2020.Month == 4) 
                           | (month_2020.Month == 5)
                           | (month_2020.Month == 6)]

Apr_June_2019 = month_2019[(month_2019.Month == 4) 
                           | (month_2019.Month == 5)
                           | (month_2019.Month == 6)]

Apr_June_2018 = month_2018[(month_2018.Month == 4) 
                           | (month_2018.Month == 5)
                           | (month_2018.Month == 6)]

Apr_June_traffic_2020 = yearly_traffic(Apr_June_2020)
Apr_June_traffic_2019 = yearly_traffic(Apr_June_2019)
Apr_June_traffic_2018 = yearly_traffic(Apr_June_2018)
print(f'2020 Apr-June Traffic: {Apr_June_traffic_2020}')
print(f'2019 Apr-June Traffic: {Apr_June_traffic_2019}')
print(f'2018 Apr-June Traffic: {Apr_June_traffic_2018}')
```

```python
# calculate Year over Year % changes for Jan-June of each year

def percent_change(current, previous):
    change_per = ((float(current)-previous)/previous)*100
    return change_per

print(f'% Change Jan - June 2019 to 2020: {round(percent_change(ttl_traffic_2020,ttl_traffic_2019),2)}')
print(f'% Change Jan - June 2018 to 2020: {round(percent_change(ttl_traffic_2020,ttl_traffic_2018),2)}')
print(f'% Change Jan - June 2018 to 2019: {round(percent_change(ttl_traffic_2019,ttl_traffic_2018),2)}')
```

```python
# calculate Year over Year % changes for April-June of each year

def percent_change(current, previous):
    change_per = ((float(current)-previous)/previous)*100
    return change_per

print(f'% Change Apr-June 2019 to 2020: {round(percent_change(Apr_June_traffic_2020,Apr_June_traffic_2019),2)}')
print(f'% Change Apr-June 2018 to 2020: {round(percent_change(Apr_June_traffic_2020,Apr_June_traffic_2018),2)}')
print(f'% Change Apr-June 2018 to 2019: {round(percent_change(Apr_June_traffic_2019,Apr_June_traffic_2018),2)}')
```

## OVERLAPPING TRAFFIC WITH NYC MILESTONE EVENTS

```python
week_2020['WeekOf'] = pd.to_datetime(week_2020['WeekOf'])
Feb_to_June = week_2020[(week_2020['WeekOf'].dt.month == 2)
          |(week_2020['WeekOf'].dt.month == 3) 
          |(week_2020['WeekOf'].dt.month == 4) 
          |(week_2020['WeekOf'].dt.month == 5) 
          |(week_2020['WeekOf'].dt.month == 6)].reset_index(drop = True)
```

```python
import matplotlib.dates as mdates

fig, ax = plt.subplots(figsize=(10,4))

ax.scatter(Feb_to_June.WeekOf, Feb_to_June.TotalTraffic)
ax.plot(Feb_to_June.WeekOf, Feb_to_June.TotalTraffic, alpha = 0.4)

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=14))
ax.set_yticklabels(["0","2M","4M","6M","8M","10M"])
plt.xlabel("WEEK", size=10, family='arial',fontsize = 12)
plt.ylabel("TRAFFIC", size=10, family='arial',fontsize = 12)
plt.title('Weekly Traffic From 02/03/2020 to 06/29/2020', size=17, y=1.02)

ax.annotate("First COVID case in New York",
            (mdates.date2num(Feb_to_June.WeekOf[3]), Feb_to_June.TotalTraffic[3]),
            xytext=(30, 0), textcoords='offset points', arrowprops=dict(arrowstyle='->'))

ax.annotate("New York declared State of Emergency",
            (mdates.date2num(Feb_to_June.WeekOf[4]), Feb_to_June.TotalTraffic[4]),
            xytext=(30, 0), textcoords='offset points', arrowprops=dict(arrowstyle='->'))

ax.annotate("Trump declared national emergency",
            (mdates.date2num(Feb_to_June.WeekOf[5]), Feb_to_June.TotalTraffic[5]),
            xytext=(30, 0), textcoords='offset points', arrowprops=dict(arrowstyle='->'))

ax.annotate("New York State on PAUSE executive order",
            (mdates.date2num(Feb_to_June.WeekOf[6]), Feb_to_June.TotalTraffic[6]),
            xytext=(-80, 20), textcoords='offset points', arrowprops=dict(arrowstyle='->'))

ax.annotate("BLM protests",
            (mdates.date2num(Feb_to_June.WeekOf[16]), Feb_to_June.TotalTraffic[16]),
            xytext=(-100, 20), textcoords='offset points', arrowprops=dict(arrowstyle='->'))

ax.annotate("Stay-at-home restrictions began to ease",
            (mdates.date2num(Feb_to_June.WeekOf[18]), Feb_to_June.TotalTraffic[18]),
            xytext=(-85, 50), textcoords='offset points', arrowprops=dict(arrowstyle='->'))

plt.savefig('nyc_milestones_2020.jpg', bbox_inches="tight")
plt.savefig('nyc_milestones_2020.svg', bbox_inches="tight");
```
